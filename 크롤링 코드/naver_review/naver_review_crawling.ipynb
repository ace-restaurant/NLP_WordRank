```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import re
import json
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
driver = webdriver.Chrome(executable_path="C:/Users/Yoon/Downloads/chromedriver.exe")
df = pd.read_csv("C:/Users/Yoon/Desktop/capstone/csv/리스트 작성 완료/강남구_주점.csv")
# null값 처리 완료 데이터 
```

```python
for i in range(len(df)):
    print('======================================================')
    print(str(i)+'번째 식당')
    # 식당 리뷰 개별 url 접속
    driver.get(df['URL'][i])
    thisurl = df['URL'][i]
    time.sleep(2) 
    # 더보기 버튼 다 누르기
    # 더보기 버튼 누르기까지 10개
    # 더보기 버튼 누르면 10개 추가됨
    # #app-root > div > div > div > div:nth-child(7) > div:nth-child(2) > div.place_section.cXO6M > div._2kAri > a
    while True:
        try:
            time.sleep(1)
            driver.find_element(by=By.TAG_NAME, value='body').send_keys(Keys.END)
            time.sleep(5)
            driver.find_element(by=By.CSS_SELECTOR, value='#app-root > div > div > div > div:nth-child(7) > div:nth-child(2) > div.place_section.cXO6M > div._2kAri > a').send_keys(Keys.ENTER)
            time.sleep(5)
            driver.find_element(by=By.TAG_NAME, value='body').send_keys(Keys.END)
            time.sleep(1)
        except NoSuchElementException:
            print('-더보기 버튼 모두 클릭 완료-')
            break
    # 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    time.sleep(1)
    # 식당 구분
    restaurant_name = df['상호명'][i]
    print('식당 이름 : '+restaurant_name)
    review_list = []
    try:
        restaurant_classification = soup.find_all('span',attrs = {'class':'_3ocDE'})[0].text
    except:
        restaurant_classification = 'none'
    df['소분류'][i] = restaurant_classification
    print('식당 구분 : '+restaurant_classification)
    print('----------------------------------------------')
    try:
        one_review = soup.find_all('div', attrs = {'class':'_3vfQ6'})
        review_num = len(one_review)
        # 특정 식당의 리뷰 총 개수
        print('리뷰 총 개수 : '+str(review_num))
        print('\n')
        # 리뷰 개수
        for i in range(review_num):
            # review 내용
            try:
                review_content = one_review[i].find('span', attrs = {'class':'WoYOw'}).text
            except:
                # 리뷰가 없다면
                review_content = ""
            print(review_content)
            print('----------------------------------------------')
            print('\n')
            #리뷰정보
            # review_list
            val = review_content.find('\n')
            if val != -1:
                review_content = review_content.replace('\n', '')
            review_content = review_content +'\t10'
            review_list.append(review_content)
    except NoSuchElementException:
        none_review = "네이버 리뷰 없음"
        print(none_review)
        review_num = 0
        print('\n')
    with open(f"{restaurant_name}.txt",'w', encoding='utf-8') as f:
        for rev in review_list:
            f.write(rev+'\n')
```
