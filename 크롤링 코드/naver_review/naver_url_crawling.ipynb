```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import re
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
df = pd.read_csv("C:/Users/Yoon/Desktop/capstone/csv/강남구_음식점_use.csv", encoding='CP949')
df['naver_map_url'] = ''  # 미리 url 담을 column 생성
driver = webdriver.Chrome(executable_path="C:/Users/Yoon/Downloads/chromedriver.exe")
```

```python
for i, keyword1 in enumerate(df['상호명']):
    print("이번에 찾을 키워드 :", i, f"/ {df.shape[0]} 행", keyword1)
    keyword2 = df['도로명주소'][i]
    try:
        naver_map_search_url = f'https://map.naver.com/v5/search/{keyword1}{keyword2}/place'
        driver.get(naver_map_search_url)
        time.sleep(4)
        cu = driver.current_url
        # 예시) https://map.naver.com/v5/search/%EA%B3%A0%ED%96%A5%EC%83%9D%EB%A7%89%EA%B1%B8%EB%A6%AC%20%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%20%EA%B0%95%EB%82%A8%EA%B5%AC%20%EC%95%95%EA%B5%AC%EC%A0%95%EB%A1%9C%20216/place/1496470647?c=14140566.2962202,4512972.0291950,15,0,0,0,dh
        res_code = re.findall(r"place/(\d+)", cu) # r-string: place/뭉쳐있는 정수 찾아서 res_code에 넣기
        final_url = 'https://pcmap.place.naver.com/restaurant/'+res_code[0]+'/review/visitor#'
        print(final_url)
        df['naver_map_url'][i] = final_url
    except IndexError:
        df['naver_map_url'][i] = ''
        print('none')
df.to_csv('url_completed.csv', encoding = 'utf-8-sig')
```
